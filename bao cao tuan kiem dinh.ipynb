```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
my_file = r'C:\Users\dcdie\Desktop\bao cao op.xlsx'
```

```python
df =pd.read_excel(my_file)
```

```python
df.shape
```

```python
dtkd = df.loc[df['lan_kd'] == 'L3'].groupby(['vu_sx','vsx','giong']) \
    .agg({'dien_tich':['count','sum']}).reset_index(names=['vu_sx','vsx','giong']).rename(columns={'count':'so_ruong_bd','sum':'dtkd_bandau'})
dtkd
```

```python
dat_ld = df.loc[(df['lan_kd'] == 'L3') & (df['ket_luan']== 'Dat')].groupby(['vu_sx','vsx','giong']) \
    .agg({'dien_tich':['count','sum']}).reset_index(names=['vu_sx','vsx','giong']).rename(columns={'count':'so_ruong_1','sum':'dt_dat_ld'})
dat_ld
```

```python
tai_kiem= df.loc[df['lan_kd'] != 'L3'].groupby(['vu_sx','vsx','giong']) \
    .agg({'dien_tich':['count','sum']}).reset_index(names=['vu_sx','vsx','giong']).rename(columns={'count':'ruong_tk','sum':'dt_tai_kiem'})
```

```python
dat_cuoi= df.loc[df['ket_luan']=='Dat'].groupby(['vu_sx','vsx','giong']) \
    .agg({'dien_tich':['count','sum']}).reset_index(names=['vu_sx','vsx','giong']).rename(columns={'count':'so_dat_cuoi','sum':'dt_dat_cuoi'})
```

```python
kq = pd.merge(dtkd, dat_ld, how= 'left', on=['vu_sx','vsx','giong'])
kq
```

```python
kq = kq.merge(tai_kiem, how= 'left', on=['vu_sx','vsx','giong'])
```

```python
kq = kq.merge(dat_cuoi, how= 'left', on=['vu_sx','vsx','giong'])
```
